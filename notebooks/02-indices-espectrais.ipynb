# 02 | Índices espectrais

Neste notebook, a cena sintética é usada para calcular NDVI e NDWI.

```python
from pathlib import Path
import sys

import matplotlib.pyplot as plt
import numpy as np

ROOT = Path.cwd().resolve().parent
if str(ROOT) not in sys.path:
    sys.path.append(str(ROOT))

from scripts.rs_utils import compute_ndvi, compute_ndwi, make_synthetic_scene, plot_single_band
```

```python
scene = make_synthetic_scene(seed=21)
ndvi = compute_ndvi(scene["nir"], scene["red"])
ndwi = compute_ndwi(scene["green"], scene["nir"])
```

```python
plot_single_band(ndvi, title="NDVI sintético", cmap="YlGn")
plot_single_band(ndwi, title="NDWI sintético", cmap="Blues")
```

```python
vegetacao = ndvi > 0.35
agua = ndwi > 0.25

fig, axes = plt.subplots(1, 2, figsize=(11, 5))
axes[0].imshow(vegetacao, cmap="Greens")
axes[0].set_title("Máscara simples de vegetação")
axes[0].set_axis_off()

axes[1].imshow(agua, cmap="Blues")
axes[1].set_title("Máscara simples de água")
axes[1].set_axis_off()
plt.tight_layout()
```

```python
print("NDVI médio:", np.round(ndvi.mean(), 3))
print("NDWI médio:", np.round(ndwi.mean(), 3))
```
