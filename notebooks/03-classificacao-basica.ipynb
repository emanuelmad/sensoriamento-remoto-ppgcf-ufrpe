# 03 | Classificação básica

Exemplo introdutório de classificação supervisionada usando uma cena sintética e `RandomForestClassifier`.

```python
from pathlib import Path
import sys

import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import train_test_split

ROOT = Path.cwd().resolve().parent
if str(ROOT) not in sys.path:
    sys.path.append(str(ROOT))

from scripts.rs_utils import make_synthetic_scene, stack_features
```

```python
scene = make_synthetic_scene(seed=11)
band_names = ["blue", "green", "red", "nir", "swir1"]
X = stack_features(scene, band_names)
y = scene["landcover"].ravel()
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)

model = RandomForestClassifier(n_estimators=120, random_state=42)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred, target_names=["água", "vegetação", "solo"]))
```

A linha acima deve imprimir a matriz de confusão e um relatório simples de desempenho.

```python
pred_map = model.predict(X).reshape(scene["landcover"].shape)

fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(scene["landcover"], cmap="viridis")
axes[0].set_title("Referência sintética")
axes[0].set_axis_off()

axes[1].imshow(pred_map, cmap="viridis")
axes[1].set_title("Mapa classificado")
axes[1].set_axis_off()
plt.tight_layout()
```
